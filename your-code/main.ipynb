## Bernoulli's Distribution

In a basket containing 100 fruits, there are 60 apples and 40 oranges. 

A fruit is picked randomly from the basket. 

1. **What is the probability that the fruit is apple?**
1. **What is the probability that it is an orange?**

````python
"""
Calculate:
p = probability that the fruit is an apple 
q = probability that the fruit is an orange
"""

import numpy as np

# Total number of fruits
total = 100
apples = 60
oranges = 40

# Probability of success (apple)
p = apples / total

# Probability of failure (orange)
q = 1 - p  # or oranges / total

print(f"p = {p:.3f}")
print(f"q = {q:.3f}")
````

Now we take a random sample of 20 fruits from the basket. After each fruit is taken, a new fruit of the same type is replaced in the basket. Therefore, every time we are taking 1 fruit from 100 fruits. 

1. **What is the probability that the first 5 fruits are all apples?**

1. **What is the probability that the first 5 fruits are all apples and the next 15 fruits are all oranges?**

You can include the `p` and `q` probabilities you previous calculated in your solution.

````python
# What is the probability that the first 5 fruits are all apples?
p = 0.6
prob_5_apples = p**5
print(round(prob_5_apples, 3))

# What is the probability that the first 5 fruits are all apples and the next 15 fruits are all oranges?
q = 0.4
prob_sequence = (p**5) * (q**15)
print(f"{prob_sequence:.3e}") # Using scientific notation for very small numbers
````

## Binomial Distribution

Similar to the above problem, this time again we randomly sample 20 fruits.

**What is the probability that this time the sample contains 5 apples and 15 oranges?**

Please note that this time the order of the fruits being picked does not matter.

````python
from scipy import stats

# Parameters
n = 20
k = 5
p = 0.6

# Calculate the Probability Mass Function (PMF)
prob_binomial = stats.binom.pmf(k, n, p)

print(f"Probability (5 apples in 20): {prob_binomial:.3f}")
````

````python
"""Final Answer:
The probability that a sample of 20 fruits contains exactly 5 apples and 15 oranges (in any order) is $0.001$.
(Note: This probability is very low because the expected number of apples in a sample of 20, given a 60% chance, is 12. 
Drawing only 5 apples is a statistically rare event.)"""
````

In the lesson we also discussed CDF (cumulative distribution function). In the context of the above problem, **what is the probability that less than 5 fruits picked are apples?**

Write the formula and calculate the probability below.

````python
from scipy import stats

# Parameters
n = 20
p = 0.6
k_less_than_5 = 4  # "Less than 5" means 0, 1, 2, 3, or 4

# Calculate CDF
prob_less_than_5 = stats.binom.cdf(k_less_than_5, n, p)

print(f"Probability of less than 5 apples: {prob_less_than_5:.3f}")

"""Final Answer:The probability that less than 5 fruits picked are apples is $0.000$ (when rounded to three decimal places).
Insight: The actual value is approximately $0.0003$. 
Because the probability of an apple is quite high (60%), it is extremely unlikely to pick 20 fruits and find 4 or fewer apples."""
````

**Plot the Probability Mass Function (PMF) for the above binomial distribution.**

*Hint: The range of x in a binomial distribution is from `0` to `n`.*

````python
# your code here
# Please label the axes and give a title to the plot 

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Parameters
n = 20
p = 0.6
x = np.arange(0, n + 1) # Range of successes from 0 to 20

# Calculate the PMF for each x
pmf_values = stats.binom.pmf(x, n, p)

# Create the plot
plt.figure(figsize=(10, 6))
plt.bar(x, pmf_values, color='skyblue', edgecolor='black', alpha=0.8)

# Formatting
plt.title(f'Binomial Distribution PMF ($n={n}, p={p}$)')
plt.xlabel('Number of Successes (Apples)')
plt.ylabel('Probability')
plt.xticks(x) # Ensure every integer is shown on the x-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.savefig('binomial_pmf.png')


"""Understanding the PlotThe Peak (Mode): The highest bar is at $x = 12$. This makes sense because the expected value (mean) 
of a binomial distribution is $n \times p = 20 \times 0.6 = 12$.Symmetry: While the distribution looks somewhat like a "Bell Curve," 
it is slightly skewed because $p \neq 0.5$. Since $p = 0.6$ is greater than $0.5$, the "bulk" of the probability is shifted toward 
the right side of the graph.Low Probabilities: Notice that the probabilities for values like $0, 1, 2, 3,$ or $4$ are almost zero. 
This confirms why our previous calculation for $P(X < 5)$ resulted in such a small number ($0.000$ when rounded).
Discrete Nature: We use a bar chart instead of a smooth line because you cannot pick $12.5$ apples; the outcomes must be integers."""
````

## Poisson Distribution

In this challenge you are required to use the Math library in python to solve the problems.

In the UEFA champions league tournament, the average number of goals scored in a match is 2.3. 

**What is the probability that the number of goals scored in a match is 5?**

*Hint: Use the exp() in the math library*

And to calculate the factorial, you can use the following code:

```python
import math
math.factorial(n) # n is the number you want to find the factorial of
```

````python
import math

# Parameters
lam = 2.3
k = 5

# Poisson Formula: (lam^k * e^-lam) / k!
probability = (math.pow(lam, k) * math.exp(-lam)) / math.factorial(k)

print(f"The probability of exactly 5 goals is: {round(probability, 3)}")
````

**Draw a poisson probability distribution plot for the number of goals to be between 0 to 10.**

````python
# your code here
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Average number of goals
lam = 2.3
# Range of goals to plot (0 to 10)
k_values = np.arange(0, 11)

# Calculate Poisson PMF for each value
poisson_probs = poisson.pmf(k_values, lam)

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(k_values, poisson_probs, color='mediumseagreen', edgecolor='black', alpha=0.8)

# Add labels and title
plt.title(f'Poisson Distribution for Goals per Match ($\lambda=2.3$)', fontsize=14)
plt.xlabel('Number of Goals', fontsize=12)
plt.ylabel('Probability', fontsize=12)
plt.xticks(k_values)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Save and show results
plt.savefig('poisson_distribution.png')
# Please label the axes and give a title to the plot 


"""Key Observations:

Most Likely Outcomes: The probabilities for 2 goals ($\approx 0.265$) and 1 goal ($\approx 0.231$) are the highest, 
reflecting that the average ($\lambda = 2.3$) falls between these two values.

Right Skew: The distribution has a "tail" extending to the right. While the average is low, 
there is still a non-zero (though very small) probability of seeing 7, 8, or more goals in a match.

Discrete Data: Like the Binomial distribution, we use bars because goals must be whole numbers."""
````
